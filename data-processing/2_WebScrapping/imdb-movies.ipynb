```python
import requests  
from bs4 import BeautifulSoup 
import pandas as pd

url = "https://www.imdb.com/chart/top/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
}

r = requests.get(url, headers=headers)  # Send a GET request to the URL and get the HTML response
soup = BeautifulSoup(r.text, "html.parser")

bs_titles = soup.select("div.cli-title h3.ipc-title__text")
bs_titles = bs_titles[:25]
titles = [title.get_text(strip=True) for title in bs_titles]


bs_ratings = soup.select("span.ckqiVr span.ipc-rating-star--rating")
bs_ratings = bs_ratings[:25]
ratings = [float(rating.get_text(strip=True)) for rating in bs_ratings]
```

```python
type(ratings[0])
```

```python
df = pd.DataFrame({
    "Title": titles,
    "Rating": ratings
})
print(df.count())
print(df.head(10))
```

```python
df.to_csv("top25_movies.csv", index=False)
```
